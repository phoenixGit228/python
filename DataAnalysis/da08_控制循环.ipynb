# if...else...

```python
x = -1
```

```python
if x< 0:
    print('x 小于 0')
else:
    print('x 不是负数')
```

```python
print('x 小于0') if x < 0 else print('x 不是负数')
```

```python
score = int(input('请输入你的分数:'))
```

```python
if score < 60:
    print('不及格')
elif 60 <= score < 70:
    print('成绩及格')
elif 70 <= score < 80:
    print('成绩良好')
elif 80 <= score < 90:
    print('成绩良好')
else:
    print('成绩优秀')
```

# range()

```python
for i in range(1,10,2):
    print(i)
```

```python
total = 0
for i in range(1,11):
    total += i
print(total)
```

# while循环

```python
x = 1
total = 0
while x <= 100:
    total += x
    x +=1
    
print(total)
```

# break

```python
s = 0
for i in range(20):
    s += i
    if i  >= 10:
        break #跳出本次循环

print(s)
```

# continue

```python
s = 0
for i in range(20):
    s += i
    if i  >= 10:
        continue # 跳出本次循环

print(s)
```

# pass

```python
s = 0
for i in range(20):
    s += i
    if i  >= 10:
        pass # 什么都不做

print(s)
```

# 列表推导式

[表达式 for x in range(x)]

```python
a = [x ** 2 for x in range(10)]
```

```python
a
```

[元素  表达式  条件]

```python
vect = [-9,5,8,7,-2,6,3,-2,1,4]
[x ** 0.5 for x in vect if x >0]
```

# try... except...

用于处理异常

```python
num = [9,8,-7,4,6,3,-2,'one']
total = 0
for i in num:
    try:
        total += i
    except:
        pass
    
print(total)
```
