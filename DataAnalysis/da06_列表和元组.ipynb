# 集合与字典

```python
a_set = {1,2,3,4,5}
b_set = {2,3,4,7,8,9}
```

```python
# 交集a_set减去两者交集
a_set - b_set
```

```python
# 交集a_set减去两者交集
a_set.difference(b_set)
```

```python
# 并集
a_set | b_set
```

```python
# 并集
a_set.union(b_set)
```

```python
# 交集
a_set & b_set
```

```python
# 交集
a_set.intersection(b_set)
```

```python
# 异或
a_set ^ b_set
```

```python
# 异或
a_set.symmetric_difference(b_set)
```

```python
a_set.add(10)
print(a_set)
```

```python
a_tuple = (1,2,3)
b_tuple = (2,3,4,5)
c_list = list(a_tuple + b_tuple)
c_list
```

```python
# 判断a_set 是否为b_set的子集
a_set < b_set
```
